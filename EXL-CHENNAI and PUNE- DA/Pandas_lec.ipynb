## Problem -Solving using Pandas 

```python
data = {
   'name': ['James', 'Billy', 'Bob', 'Amy', 'Tom'],
   'number': ['1234', '5678', '2222', '1111', '0909']
}
```

```python
import pandas as pd
people=pd.DataFrame(data,index=data['name'])
people
```

```python
people.axes
```

```python
name=people.loc[input("enter name:")]
print(name)
```

```python
name=input("enter name:")
print(people.loc[name])
```

```python
import pandas as pd
people=pd.DataFrame(data,index=data['name'])
people
name=people.loc[input()]
print(name)
```

```python
import pandas as pd
people=pd.DataFrame(data,index=data['name'])
people
start=int(input("enter the starting record number:"))
end=int(input("enter the ending record number:"))
name=people.iloc[start:end]
print(name)
```
